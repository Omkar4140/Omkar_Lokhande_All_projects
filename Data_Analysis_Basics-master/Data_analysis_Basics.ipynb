```python
import pandas as pd
df=pd.read_csv(r'/content/uncleaned_appliances (1).csv')
#change column name
df.rename(columns={'goods-title-link--jump':'GTLJ','goods-title-link--jump href':'GTLJH','selling_proposition':'SP','goods-title-link':'GTL'},inplace=True)
#fillna method to deal with NAN values
df['GTLJ']=df['GTLJ'].fillna(method='ffill')
df['GTLJH']=df['GTLJH'].fillna(method='ffill')
df['SP']=df['SP'].fillna(method='backfill')
df['SP']=df['SP'].fillna(method='ffill')
df['GTL']=df['GTL'].fillna(method='backfill')
df['rank-title']=df['rank-title'].fillna(method='ffill')
df['rank-sub']=df['rank-sub'].fillna(method='ffill')
df['price']=df['price'].fillna(0)
df['discount']=df['discount'].fillna(0)
```

```python
#remove duplicate values
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(1,6)
a
b=a**2
b
plt.plot(a,b,color='Green',marker='*',markersize=20,markerfacecolor='red',markeredgecolor='yellow',linewidth=3,
         linestyle='dotted',label='sq_num')
plt.title('sq')
plt.xlabel('x')
plt.ylabel('y')
c=a**3
plt.plot(c,marker='o',label='cube')
plt.plot(loc='center')
```
