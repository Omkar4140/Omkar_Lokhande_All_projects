```python
import matplotlib.pyplot as plt

labels = ['Skill-Based', 'Traditional']
rates = [85, 65]

plt.bar(labels, rates, color=['#4CAF50', '#FF5722'])
plt.title('Employment Rates by Education Type')
plt.ylabel('Employment Rate (%)')
plt.show()
```

```python
industries = ['IT', 'Healthcare', 'Manufacturing', 'Retail']
rates = [88, 82, 78, 65]

plt.barh(industries, rates, color='#2196F3')
plt.title('Employability by Industry (Skill-Based Graduates)')
plt.xlabel('Employability Rate (%)')
plt.show()
```

```python
import numpy as np

labels = ['Skill-Based', 'Traditional']
salaries = [45000, 38000]

x = np.arange(len(labels))
plt.bar(x, salaries, color=['#009688', '#E91E63'])
plt.xticks(x, labels)
plt.title('Average Starting Salaries')
plt.ylabel('Salary ($)')
plt.show()
```

```python
4. Job Readiness Perception (Stacked Bar Chart)
Data Table:

Education Type	Highly Prepared (%)	Moderately Prepared (%)	Unprepared (%)
Skill-Based	72	20	8
Traditional	45	35	20
```

```python
labels = ['Skill-Based', 'Traditional']
high = [72, 45]
moderate = [20, 35]
unprepared = [8, 20]

plt.bar(labels, high, label='Highly Prepared', color='#4CAF50')
plt.bar(labels, moderate, bottom=high, label='Moderately Prepared', color='#FFC107')
plt.bar(labels, unprepared, bottom=np.add(high, moderate), label='Unprepared', color='#F44336')
plt.ylabel('Percentage (%)')
plt.title('Job Readiness Perception')
plt.legend()
plt.show()
```

```python
labels = ['Satisfied', 'Neutral', 'Dissatisfied']
sizes = [78, 15, 7]
colors = ['#4CAF50', '#FFC107', '#F44336']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Employer Satisfaction')
plt.show()
```

```python
labels = ['Aligned', 'Misaligned']
sizes = [75, 25]
colors = ['#2196F3', '#FF5722']

plt.pie(sizes, labels=labels, colors=colors, wedgeprops={'width': 0.4}, autopct='%1.1f%%')
plt.title('Curriculum Alignment with Industry Needs')
plt.show()
```
